# Main Function to Analyze the images

```python
import os

#    https://github.com/bsairesearch/faceconstruct/blob/main/faces/andro.jpeg
#loc = '/Users/suresh/dev/FaceConstruct/faceconstruct/faces/'
def get_imageanalysis(lst, loc=None):
    lst_moves = []
    lst_conf = []
    print("\n\nLocation = " + loc)
    i = 0
    #for filename in os.listdir(loc):
    for filename in lst:
        print(str(i) + ". " + loc + filename)
        i = i + 1
        sleep(2)
        with io.open(loc + filename, 'rb') as image_file:
            content = image_file.read()        
            image = vision.Image(content=content)
            response = client.face_detection(image=image)
            faces = response.face_annotations
            print(filename +"==" + str(faces))
            conf_score=-1
            if len(faces) > 0:
                conf_score = round(faces[0].detection_confidence * 100)
            lst_conf.append(filename +"==" + str(conf_score))
            lst_moves.append('mv ' + loc_FF + filename + '  ' + loc_FF_host)
            
            print(filename+"\n############\n")
            #print(filename +"==" + str(faces))
        
    print("\n############\n")
    print(lst_conf)
    print(*lst_conf, sep = "\n")

    print("\n############\n")
    print(lst_moves)
    print(*lst_moves, sep = "\n")    
    
    print('\n***DONE***\n')
```
